```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def sigmoid_deriv(x):
    return sigmoid(x) * (1 - sigmoid(x))

def relu_deriv(x):
    return np.where(x > 0, 1, 0)

def tanh_deriv(x):
    return 1 - np.square(tanh(x))

def elu_deriv(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

x = np.linspace(-3, 3, 1000)

# Plot activation functions
plt.figure(figsize=(12, 6))
plt.plot(x, sigmoid(x), label='Sigmoid', ls='solid')
plt.plot(x, relu(x), label='ReLU', ls='dotted')
plt.plot(x, tanh(x), label='Tanh', ls='dashed')
plt.plot(x, elu(x), label='ELU', ls='dashdot')
# plt.title('Activation Functions')
plt.legend()
plt.grid()
plt.savefig(fig_path+'Activation.png')
# Plot activation functions' derivatives
plt.figure(figsize=(12, 6))
plt.plot(x, sigmoid_deriv(x), label='Sigmoid Derivative', ls='solid')
plt.plot(x, relu_deriv(x), label='ReLU Derivative', ls='dotted')
plt.plot(x, tanh_deriv(x), label='Tanh Derivative', ls='dashed')
plt.plot(x, elu_deriv(x), label='ELU Derivative', ls='dashdot')
# plt.title("Activation Functions' Derivatives")
plt.legend()
plt.grid()
plt.savefig(generateFilePath('Derivatives'))
```
