```python
def calculate_annual_spending(monthly_spending_list):
    """
    计算年度消费总额
    参数：monthly_spending_list - 包含12个月消费额的列表
    返回：年度消费总额（浮点数）
    """
    # 检查输入是否为12个月的数据
    if len(monthly_spending_list) != 12:
        print("错误：输入必须包含12个月的消费数据")
        return 0  # 输入异常时返回0
    annual_total = sum(monthly_spending_list)  # 求和
    return round(annual_total, 2)  # 保留两位小数

# 调用函数：计算某客户的年度消费
customer_monthly = [3200, 2800, 3100, 2900, 3500, 4200, 3800, 3600, 3300, 3000, 2700, 4100]
annual = calculate_annual_spending(customer_monthly)
print(f"年度消费总额：{annual}元")  # 输出：年度消费总额：39200元
```

```python
def get_customer_level(annual_spending):
    if annual_spending >= 100000:
        return "钻石会员"
    elif annual_spending >= 50000:
        return "黄金会员"
    elif annual_spending >= 20000:
        return "白银会员"
    else:
        return "普通会员"

# 循环处理多个客户的消费数据
customers_annual = [120000, 65000, 30000, 15000]
for spending in customers_annual:
    level = get_customer_level(spending)
    print(f"年度消费{spending}元 → {level}")
```

```python
class Customer:
    # 初始化方法：创建对象时自动调用，定义属性
    def __init__(self, cid, name, initial_balance=0):
        self.cid = cid  # 客户ID
        self.name = name  # 姓名
        self.balance = initial_balance  # 账户余额
        self.purchase_history = []  # 购买记录（空列表）
    
    # 方法：添加购买记录并扣减余额
    def make_purchase(self, product, price):
        self.purchase_history.append(product)  # 记录购买商品
        self.balance -= price  # 扣减余额
        print(f"{self.name}购买了{product}，花费{price}元，剩余余额：{self.balance}元")
    
    # 方法：查询购买历史
    def get_history(self):
        return f"{self.name}的购买记录：{self.purchase_history}"

# 创建客户对象
customer1 = Customer("C001", "王五", 10000)  # 初始余额10000元
customer2 = Customer("C002", "赵六")  # 初始余额默认0元

# 调用对象方法
customer1.make_purchase("笔记本电脑", 6999)  # 输出：王五购买了笔记本电脑，花费6999元，剩余余额：3001元
customer1.make_purchase("鼠标", 99)  # 输出：王五购买了鼠标，花费99元，剩余余额：2902元
print(customer1.get_history())  # 输出：王五的购买记录：['笔记本电脑', '鼠标']
```

```python
class VIPCustomer(Customer):
    # 重写初始化方法，增加折扣率
    def __init__(self, cid, name, initial_balance=0, discount=0.9):
        super().__init__(cid, name, initial_balance)  # 调用父类初始化方法
        self.discount = discount  # 折扣率（默认9折）
    
    # 重写购买方法，应用折扣
    def make_purchase(self, product, price):
        discounted_price = price * self.discount  # 计算折扣后价格
        super().make_purchase(product, discounted_price)  # 调用父类方法扣减余额

# 创建VIP客户对象
vip_customer = VIPCustomer("V001", "钱七", 20000, 0.8)  # 8折优惠
vip_customer.make_purchase("手机", 5000)  # 输出：钱七购买了手机，花费4000.0元，剩余余额：16000.0元
```

```python
# 模拟含异常值的销售数据
sales = pd.DataFrame({
    "产品": ["A", "A", "A", "B", "B", "B"],
    "销量": [120, 135, 140, 85, 90, 300]  # 300为异常值（远高于其他值）
})

# 1. 描述性统计识别异常值：查看最大值、四分位距（IQR）
print("销量描述性统计：")
print(sales["销量"].describe())
# 输出中max=300明显高于75%分位数140，可能为异常

# 2. 用IQR法定义异常值阈值（低于Q1-1.5*IQR或高于Q3+1.5*IQR）
Q1 = sales["销量"].quantile(0.25)
Q3 = sales["销量"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 筛选异常值
outliers = sales[(sales["销量"] < lower_bound) | (sales["销量"] > upper_bound)]
print("\n异常值：")
print(outliers)  # 输出销量为300的记录

# 3. 处理异常值：替换为边界值或删除
sales_cleaned = sales.copy()
# 将异常值替换为上边界
sales_cleaned["销量"] = sales_cleaned["销量"].apply(
    lambda x: upper_bound if x > upper_bound else x
)
print("\n处理后的销量：")
print(sales_cleaned["销量"])
```
